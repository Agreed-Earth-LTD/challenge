## Introduction

This notebook compares field boundaries provided by Agreed Earth and Agrimetrics.
The field polygons are visualized in their geographic context along with their differences.

## Imports

```python
import matplotlib.pyplot as plt
import pandas as pd
import geopandas as gp
from shapely import wkt
import contextily as cx
```

```python
overlapping_data = '../data/overlapping.csv'
```

```python
overlap = pd.read_csv(overlapping_data)
```

```python
overlap.head()
```

```python
type(overlap)
```

Pandas DataFrame to then use `wkt.loads` FTW courtesy of [SO](https://stackoverflow.com/questions/61122875/geopandas-how-to-read-a-csv-and-convert-to-a-geopandas-dataframe-with-polygons).

```python
overlap['Agrimetrics Geometry'] = overlap['Agrimetrics Geometry'].apply(wkt.loads)
overlap['Agreed Earth Geometry'] = overlap['Agreed Earth Geometry'].apply(wkt.loads)
overlap.head()
```

```python
overlap_gdf = gp.GeoDataFrame(overlap)
overlap_gdf.head()
```

```python
agri_overlap = overlap_gdf.set_geometry('Agrimetrics Geometry', crs='EPSG:4326')
agri_overlap.head()
```

```python
ax = agri_overlap.to_crs('EPSG:4326').plot(figsize=(20, 10))
ax.set_title('Outwoods Farm fields\n(Agrimetrics boundaries)')
ax.set_xlabel('Long.')
ax.set_ylabel('Lat.')
cx.add_basemap(ax, source=cx.providers.OpenStreetMap.Mapnik, crs='EPSG:4326')
plt.savefig('../plots/outwoods_fields.png', facecolor='white');
```

```python
# BNG is EPSG 27700
agree_overlap = overlap_gdf.set_geometry('Agreed Earth Geometry', crs='EPSG:27700')
agree_overlap.head()
```

```python
overlap['data_id'].value_counts()
```

Note some IDs are duplicated, for example 'Intake-3'.

```python
overlap[overlap.data_id == 'Intake-3']
```

```python
counts = overlap['data_id'].value_counts()
duplicates = counts[counts > 1].index.to_list()
duplicates
```

```python
ax = agree_overlap[agree_overlap.data_id.isin(duplicates)].plot(column='data_id', legend=True, figsize=(15, 15))
ax.set_title('Locations and names of duplicated fields\n(Agreed Earth boundaries)')
ax.set_xlabel('Easting')
ax.set_ylabel('Northing')
cx.add_basemap(ax, source=cx.providers.OpenStreetMap.Mapnik, crs='EPSG:27700');
```

```python
for f in overlap[overlap.data_id == 'Intake-3']['agrimetrics_id']:
    print(f)
```

In this example, the duplicated ID (field name) is associated with two different Agrimetrics IDs.

```python
ax = agree_overlap.plot(alpha=0.5, figsize=(15, 15))
ax.set_title('Outwoods Farm fields\n(weighted by frequency to show duplicates)')
ax.set_xlabel('Easting')
ax.set_ylabel('Northing')
cx.add_basemap(ax, source=cx.providers.OpenStreetMap.Mapnik, crs='EPSG:27700')
plt.savefig('../plots/outwoods_fields_Agreed.png', facecolor='white');
```

## Symmetric difference between Agrimetrics and Agreed Earth field boundaries

```python
sym_diff = agri_overlap['Agrimetrics Geometry'].to_crs('EPSG:27700').symmetric_difference(agree_overlap['Agreed Earth Geometry'])
```

```python
sym_diff.head()
```

```python
ax = sym_diff.plot(figsize=(15, 15), color='r', alpha=0.5)
ax.set_title('Symmetric difference between Agrimetrics and Agreed Earth boundaries\n(weighted by duplicates)')
ax.set_xlabel('Easting')
ax.set_ylabel('Northing')
cx.add_basemap(ax, source=cx.providers.OpenStreetMap.Mapnik, crs='EPSG:27700')
plt.savefig('../plots/outwoods_fields_sym_diff.png', facecolor='white');
```

The Agrimetrics boundaries are consistent with automated segmentation techniques, hence the boundaries being eroded from the nominal boundaries supplied by Agreed.
Furthermore, the Agreed boundaries can sometimes span apparent (in the map) boundaries, suggesting Agrimetrics is better at identifying actual boundaries between parcels of land
based on crop type or use.
This suggests Agreed will overestimate the area of land under cultivation.
